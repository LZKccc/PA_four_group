### 任务1:新建项目文件夹

- 需要使用 Python 命令自动化创建文件夹和子文件夹。
- 主文件夹：xxxx。添加如下文档：
- README.md (用于存储数据集的描述)
- 01_data_clean.ipynb (数据清洗)
- 02_data_analysis.ipynb (数据分析)
- 子文件夹：
- data_raw (用于存储下载的原始数据) 和 data_clean (用于存储清洗后的数据)
- output (用于存储输出的图形和表格)

```python
import os

# 定义项目根目录名称
project_name = "city_budget_analysis"

# 定义需要创建的子文件夹
sub_folders = [
    "data_raw",   # 存放原始 CSV 文件
    "data_clean",  # 存放清洗后的数据
    "output"      # 存放分析图表和表格
]

# 定义需要创建的文件
files = [
    "README.md",
    "01_data_clean.ipynb",
    "02_data_analysis.ipynb"
]

def setup_project():
    # 1. 创建主文件夹
    if not os.path.exists(project_name):
        os.makedirs(project_name)
        print(f"✔️ 已创建主项目文件夹: {project_name}")

    # 2. 创建子文件夹
    for folder in sub_folders:
        path = os.path.join(project_name, folder)
        if not os.path.exists(path):
            os.makedirs(path)
            print(f"  └── 已创建子文件夹: {folder}")

    # 3. 创建初始化文件
    for file in files:
        file_path = os.path.join(project_name, file)
        if not os.path.exists(file_path):
            with open(file_path, 'w', encoding='utf-8') as f:
                if file == "README.md":
                    f.write("# 中国主要城市公共预算数据分析项目\n\n## 数据集描述\n"
                            "- **city_expenditure.csv**: 地方一般公共预算支出(亿元)\n"
                            "- **city_income.csv**: 地方一般公共预算收入(亿元)\n"
                            "- **gdp.csv**: 地区生产总值(亿元)\n"
                            "- **individual_deposit.csv**: 住户存款余额(亿元)\n"
                            "- **Real_estate_development_investment.csv**: 房地产开发投资额(亿元)\n"
                            "- **Sales_area_of_houses.csv**: 商品房销售面积(万平方米)\n")
                # Jupyter Notebook 需要基本的 JSON 结构才能被识别，这里先创建为空，手动打开即可
                pass
            print(f"  └── 已创建文件: {file}")

if __name__ == "__main__":
    setup_project()
    print("\n🚀 项目结构初始化完成！请将上传的 CSV 文件放入 data_raw 文件夹中。")
```
